# <a id='toc1_'></a>[Проект для «Викишоп»](#toc0_)

Интернет-магазин «Викишоп» запускает новый сервис. Теперь пользователи могут редактировать и дополнять описания товаров, как в вики-сообществах. То есть клиенты предлагают свои правки и комментируют изменения других. Магазину нужен инструмент, который будет искать токсичные комментарии и отправлять их на модерацию.

Обучите модель классифицировать комментарии на позитивные и негативные. В вашем распоряжении набор данных с разметкой о токсичности правок.

Постройте модель со значением метрики качества *F1* не меньше 0.75.

**Описание данных**

Данные находятся в файле `toxic_comments.csv`. Столбец *text* в нём содержит текст комментария, а *toxic* — целевой признак.

## <a id='toc1_1_'></a>[Подготовка](#toc0_)

```python
import pandas as pd
import numpy as np

import re
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords, wordnet
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import lightgbm as lgb
from sklearn.metrics import f1_score

import warnings
warnings.filterwarnings('ignore')
```

### <a id='toc1_1_1_'></a>[Посмотрим на данные](#toc0_)

```python
data = pd.read_csv('./datasets/toxic_comments.csv')
data.head()
```

```python
data.info()
```

```python
data.duplicated().sum()
```

```python
data.groupby('toxic').size()
```

В данных нет пропусков и дубликатов

Есть дисбаланс классов

### <a id='toc1_1_2_'></a>[Очистим текст от лишних символов и приведем к нижнему регистру](#toc0_)

```python
def clear_text(text):
    text = re.sub(r'[^a-zA-Z ]', ' ', text)
    text = " ".join(text.split())
    text = text.lower()
    return text
```

```python
data['text'] = data['text'].apply(clear_text)
data.head()
```

### <a id='toc1_1_3_'></a>[Лемматизируем текст](#toc0_)

```python
def lemmatize(text):

    def get_wordnet_pos(word):
        tag = nltk.pos_tag([word])[0][1][0].upper()
        tag_dict = {"J": wordnet.ADJ,
                    "N": wordnet.NOUN,
                    "V": wordnet.VERB,
                    "R": wordnet.ADV}
        return tag_dict.get(tag, wordnet.NOUN)

    lemmatizer = WordNetLemmatizer()
    word_list = nltk.word_tokenize(text)
    lemm_text = [lemmatizer.lemmatize(w, get_wordnet_pos(w)) for w in word_list]
    joined = ' '.join(lemm_text)
    return joined
```

```python
data['text'] = data['text'].apply(lemmatize)
data.head()
```

### <a id='toc1_1_4_'></a>[Найдем стоп-слова](#toc0_)

```python
stopwords = list(stopwords.words('english'))
```

### <a id='toc1_1_5_'></a>[Разделим данные на обучающую и тестовую выборки](#toc0_)

```python
# запишем random_state в переменную

RS = 12345
```

```python
features = data['text']
target = data['toxic']
```

```python
features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.25, random_state=RS, stratify=target
)
print(features_train.shape)
print(features_test.shape)
```

```python
# создадим корпуса для признаков

features_train_corpus = features_train.values
features_test_corpus = features_test.values
```

### <a id='toc1_1_6_'></a>[Вывод](#toc0_)

При осмотре данных обнаружено отсутствие пропусков и дубликатов

Данные не сбалансированы, соотношение классов примерно равно 9:1

Тексты очищены от лишних символов и лемматизированы

Данные разделены на обучающую и тестовую выборки в соотношении 3:1 с учетом дисбаланса классов

## <a id='toc1_2_'></a>[Обучение](#toc0_)

### <a id='toc1_2_1_'></a>[Обучим модели с помощью мешка слов](#toc0_)

```python
count_vect = CountVectorizer(stop_words=stopwords, dtype=np.float64)
features_train_bow = count_vect.fit_transform(features_train_corpus)
features_test_bow = count_vect.transform(features_test_corpus)
```

#### <a id='toc1_2_1_1_'></a>[Обучим модель линейной регрессии](#toc0_)

```python
model_lr = LogisticRegression(class_weight='balanced', random_state=RS)
params = {'fit_intercept': [True, False]}

grid_lr = RandomizedSearchCV(model_lr,
                             params,
                             n_iter=5,
                             cv=5,
                             n_jobs=-1,
                             scoring='f1',
                             random_state=RS)
grid_lr.fit(features_train_bow, target_train)
f1_lr = grid_lr.best_score_

print('Лучшие параметры: ', grid_lr.best_params_)
print('F1: ', f1_lr)
```

#### <a id='toc1_2_1_2_'></a>[Обучим модель случайного леса](#toc0_)

```python
model_rfc = RandomForestClassifier(class_weight='balanced', random_state=RS)
params = {'max_depth': [230, 240, 250, 260, 270],
          'n_estimators': [5, 6, 7]
}

grid_rfc = RandomizedSearchCV(model_rfc,
                              params,
                              n_iter=5,
                              cv=5,
                              n_jobs=-1,
                              scoring='f1',
                              random_state=RS)
grid_rfc.fit(features_train_bow, target_train)
f1_rfc = grid_rfc.best_score_

print('Лучшие параметры: ', grid_rfc.best_params_)
print('F1: ', f1_rfc)
```

#### <a id='toc1_2_1_3_'></a>[Обучим модель LightGBM](#toc0_)

```python
model_lgb = lgb.LGBMClassifier(class_weight='balanced', random_state=RS)
params = {
    'max_depth': [3, 5, 7, 9],
    'num_iterations': [500, 600, 700, 800, 900, 1000],
    'learning_rate': [0.001, 0.01, 0.1]
}

grid_lgb = RandomizedSearchCV(estimator=model_lgb,
                              param_distributions=params,
                              n_iter=5,
                              cv=5,
                              scoring='f1',
                              n_jobs=-1,
                              random_state=RS
                             )
grid_lgb.fit(features_train_bow, target_train)
f1_lgb = grid_lgb.best_score_

print('Лучшие параметры: ', grid_lgb.best_params_)
print('F1: ', f1_lgb)
```

### <a id='toc1_2_2_'></a>[Обучим модели с помощью TF-IDF](#toc0_)

```python
tfidf_vect = TfidfVectorizer(stop_words=stopwords, dtype=np.float64)
features_train_tfidf = tfidf_vect.fit_transform(features_train_corpus)
features_test_tfidf = tfidf_vect.transform(features_test_corpus)
```

#### <a id='toc1_2_2_1_'></a>[Обучим модель линейной регрессии](#toc0_)

```python
model_lr = LogisticRegression(class_weight='balanced', random_state=RS)
params = {'fit_intercept': [True, False]}

grid_lr = RandomizedSearchCV(model_lr,
                             params,
                             n_iter=5,
                             cv=5,
                             n_jobs=-1,
                             scoring='f1',
                             random_state=RS)
grid_lr.fit(features_train_tfidf, target_train)
f1_lr = grid_lr.best_score_

print('Лучшие параметры: ', grid_lr.best_params_)
print('F1: ', f1_lr)
```

#### <a id='toc1_2_2_2_'></a>[Обучим модель случайного леса](#toc0_)

```python
model_rfc = RandomForestClassifier(class_weight='balanced', random_state=RS)
params = {'max_depth': [230, 240, 250, 260, 270],
          'n_estimators': [5, 6, 7]
}

grid_rfc = RandomizedSearchCV(model_rfc,
                              params,
                              n_iter=10,
                              cv=5,
                              n_jobs=-1,
                              scoring='f1',
                              random_state=RS)
grid_rfc.fit(features_train_tfidf, target_train)
f1_rfc = grid_rfc.best_score_

print('Лучшие параметры: ', grid_rfc.best_params_)
print('F1: ', f1_rfc)
```

#### <a id='toc1_2_2_3_'></a>[Обучим модель LightGBM](#toc0_)

```python
model_lgb = lgb.LGBMClassifier(class_weight='balanced', random_state=RS)
params = {
    'max_depth': [3, 5, 7, 9],
    'num_iterations': [500, 600, 700, 800, 900, 1000],
    'learning_rate': [0.001, 0.01, 0.1]
}

grid_lgb = RandomizedSearchCV(estimator=model_lgb,
                              param_distributions=params,
                              n_iter=5,
                              cv=5,
                              scoring='f1',
                              n_jobs=-1,
                              random_state=RS
                             )
grid_lgb.fit(features_train_tfidf, target_train)
f1_lgb = grid_lgb.best_score_

print('Лучшие параметры: ', grid_lgb.best_params_)
print('F1: ', f1_lgb)
```

### <a id='toc1_2_3_'></a>[Вывод](#toc0_)

Были обучены модели с помощью мешка слов:
- Логистическая регрессия, F1-мера ≈ 0,750
- Случайный лес, F1-мера ≈ 0,572
- LightGBM, F1-мера ≈ 0,760

С помощью TF-IDF:
- Логистическая регрессия, F1-мера ≈ 0,745
- Случайный лес, F1-мера ≈ 0,565
- LightGBM, F1-мера ≈ 0,761

Лучший результат показала модель LightGBM, обученная с помощью TF-IDF. Проверим ее на тестовой выборке.

## <a id='toc1_3_'></a>[Тестирование](#toc0_)

```python
best_model = lgb.LGBMClassifier(
    class_weight='balanced',
    num_iterations=900,
    max_depth=7,
    learning_rate=0.1,
    random_state=RS
)
best_model.fit(features_train_tfidf, target_train)
best_predictions = best_model.predict(features_test_tfidf)
best_f1 = f1_score(target_test, best_predictions)
print('F1-мера на тестовой выборке: ', best_f1)
```

Модель LightGBM, обученная с помощью TF-IDF, показала на тестовой выборке требуемый результат, F1-мера ≈ 0,762
