## Умножение матриц

В этом задании вы можете записывать формулы в *Jupyter Notebook.*

Для примера мы записали формулы линейной регрессии. Можете их скопировать и отредактировать, чтобы решить задачу.

Обозначения:

- $X$ — матрица признаков (нулевой столбец состоит из единиц)

- $y$ — вектор целевого признака

- $P$ — матрица, на которую умножаются признаки

- $w$ — вектор весов линейной регрессии (нулевой элемент равен сдвигу)

Предсказания:

$$
a = Xw
$$

Задача обучения:

$$
w = \arg\min_w MSE(Xw, y)
$$

Формула обучения:

$$
w = (X^T X)^{-1} X^T y
$$

**Ответ:** При умножении признаков на обратимую матрицу качество линейной регрессии не изменится. Предсказания $a$ равны $a_1$

**Обоснование:** Докажем, что $a = a_1$

$$
a_1 = XPw_1
$$

$$
w_1 = ((XP)^T XP)^{-1} (XP)^T y
$$

Подставим в формулу:

$$
a_1 = XP((XP)^T XP)^{-1} (XP)^T y
$$

Так как $(AB)^T = B^T A^T$:

$$
a_1 = XP(P^T(X^T X)P)^{-1} P^T X^T y
$$

Так как $(AB)^{-1} = B^{-1}A^{-1}$:

$$
a_1 = XPP^{-1} (X^T X)^{-1} (P^T)^{-1} P^T X^T y
$$

Так как $AA^{-1} = A^{-1}A = E$:

$$
a_1 = XE (X^T X)^{-1} EX^T y
$$

Так как $AE = EA = A$:

$$
a_1 = X (X^T X)^{-1} X^T y
$$

Так как $(X^T X)^{-1} X^T y = w$:

$$
a_1 = Xw
$$

$$
a_1 = a
$$

Из формулы выше:

$$
w_1 = P^{-1} (X^T X)^{-1} X^T y
$$

$$
w_1 = P^{-1}w
$$

## Алгоритм преобразования

**Алгоритм**

Мы сгенерируем квадратную матрицу размером 4х4 - по количеству столбцов признаков - проверим ее на обратимость и умножим на матрицу признаков.

Посчитаем качество модели, обученной на исходных признаках и сравним с качеством модели, обученной на зашифрованных признаках.

**Обоснование**

Мы доказали, что предсказания для моделей будут одинаковы, а значит и метрики R2 должны быть равны

## Проверка алгоритма

```python
features = data.drop('Страховые выплаты', axis=1)
target = data['Страховые выплаты']
```

Сгенерируем квадратную матрицу и проверим, что она обратима.

```python
square_matrix = np.random.normal(size=(4, 4))
square_matrix
```

```python
inverted_matrix = np.linalg.inv(square_matrix)
inverted_matrix
```

Умножим матрицу признаков на квадратную матрицу

```python
encrypted_matrix = features.values@square_matrix
encrypted_features = pd.DataFrame(encrypted_matrix, columns=features.columns)
encrypted_features
```

Обучим модель на начальных данных

```python
model = LinearRegression()
model.fit(features, target)
predictions = model.predict(features)
print(r2_score(target, predictions))
```

Теперь обучим модель на зашифрованных данных

```python
model = LinearRegression()
model.fit(encrypted_features, target)
predictions = model.predict(encrypted_features)
print(r2_score(target, predictions))
```

### Вывод

Мы видим, что метрики отличаются незначительно, а значит, данные зашифрованы правильно.

## Декодирование данных

$$
(XP)P^{-1} = XE = X
$$

Умножим матрицу с зашифрованными данными на матрицу, обратную нашей квадратной матрице 

```python
decoded_matrix = encrypted_matrix@inverted_matrix
```

```python
decoded_features = pd.DataFrame(decoded_matrix.round(0).astype(int), columns=features.columns)
decoded_features.head(10)
```

```python
decoded_features.compare(features)
```

Различий между декодированным датафреймом и исходным не обнаружено. Данные декодированы успешно.
