# Отток клиентов

Из «Бета-Банка» стали уходить клиенты. Каждый месяц. Немного, но заметно. Банковские маркетологи посчитали: сохранять текущих клиентов дешевле, чем привлекать новых.

Нужно спрогнозировать, уйдёт клиент из банка в ближайшее время или нет. Вам предоставлены исторические данные о поведении клиентов и расторжении договоров с банком. 

Постройте модель с предельно большим значением *F1*-меры. Чтобы сдать проект успешно, нужно довести метрику до 0.59. Проверьте *F1*-меру на тестовой выборке самостоятельно.

Дополнительно измеряйте *AUC-ROC*, сравнивайте её значение с *F1*-мерой.

Источник данных: [https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling](https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling)

## Подготовка данных

### Импортируем библиотеки

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
from sklearn.utils import shuffle
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.metrics import confusion_matrix
```

### Загрузим и осмотрим данные

```python
data = pd.read_csv('/datasets/Churn.csv')
```

```python
data.head()
```

```python
data.info()
```

### Удалим неинформативные столбцы

Удалим столбцы RowNumber, CustomerId, Surname

```python
data = data.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
```

### Заполним пропуски

Есть пропуски в колонке Tenure, заменим их средним значением.

```python
data['Tenure'].fillna(data['Tenure'].mean(), inplace=True)
data.info()
```

### Проверим данные на дубликаты

```python
data.duplicated().sum()
```

Явных дубликатов нет.

### Разобьем данные на выборки

Разобьем данные на тренировочную, валидационную и тестовую выборки в соотношении 3:1:1.

```python
features = data.drop(['Exited'], axis=1)
target = data['Exited']

features1, features_test_raw, target1, target_test = train_test_split(
    features, target, test_size=0.2, random_state=12345, stratify=target
) 
features_train_raw, features_valid_raw, target_train, target_valid = train_test_split(
    features1, target1, test_size=0.25, random_state=12345, stratify=target1
)

print(features_train_raw.shape[0])
print(features_valid_raw.shape[0])
print(features_test_raw.shape[0])
print()
print(target_train[target_train == 1].sum())
print(target_valid[target_valid == 1].sum())
print(target_test[target_test == 1].sum())
```

### Подготовим категориальные признаки

Закодируем категориальные признаки с помощью техники OHE.

```python
transformer = make_column_transformer((OneHotEncoder(drop='first'), ['Geography', 'Gender']), remainder='passthrough')
features_train = pd.DataFrame(transformer.fit_transform(features_train_raw),
                              columns=transformer.get_feature_names(),
                              index=features_train_raw.index
                             )
features_valid = pd.DataFrame(transformer.transform(features_valid_raw),
                              columns=transformer.get_feature_names(),
                              index=features_valid_raw.index
                             )
features_test = pd.DataFrame(transformer.transform(features_test_raw),
                             columns=transformer.get_feature_names(),
                             index=features_test_raw.index
                            )
```

### Подготовим количественные признаки

Стандартизируем количественные признаки.

```python
pd.options.mode.chained_assignment = None

numeric = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']

scaler = StandardScaler()
features_train[numeric] = scaler.fit_transform(features_train[numeric])
features_valid[numeric] = scaler.transform(features_valid[numeric])
features_test[numeric] = scaler.transform(features_test[numeric])
```

## Исследование задачи

### Изучим баланс классов в выборке.

```python
class_frequency = data['Exited'].value_counts(normalize=True)
print(class_frequency)
class_frequency.plot(kind='bar')
```

Видим, что данные не сбалансированы. Доля ушедших от банка клиентов около 20%, это далеко не 1:1.

### Обучим модель на несбалансированных данных.

#### Проверим модель дерева решений на несбалансированных данных.

```python
best_result = 0
for depth in range(1, 11):
    for feature in range(1, 11):
        for leaf in range(1, 11):
            dtc_nb = DecisionTreeClassifier(
                max_depth=depth,
                max_features=feature,
                min_samples_leaf=leaf,
                random_state=12345
            )
            dtc_nb.fit(features_train, target_train)
            predictions_valid_dtc_nb = dtc_nb.predict(features_valid)
            probabilities_one_valid_dtc_nb = dtc_nb.predict_proba(features_valid)[:, 1]
            result = f1_score(target_valid, predictions_valid_dtc_nb)
            auc_roc = roc_auc_score(target_valid, probabilities_one_valid_dtc_nb)
            matrix = confusion_matrix(target_valid, predictions_valid_dtc_nb)
            if result > best_result:
                best_result = result
                best_auc_roc = auc_roc
                best_matrix = matrix
                best_depth = depth
                best_max_features=feature
                best_min_samples_leaf=leaf
print(best_result, best_depth, best_max_features, best_min_samples_leaf)
print(best_auc_roc)
print(best_matrix)
```

Значение F1-меры для дерева решений на несбалансированных данных = 0.5971428571428572 при гиперпараметрах depth = 8, max_features = 10, min_samples_leaf = 6. Здесь мы достигли требуемого результата, но продолжим пробовать другие модели для получения наилучшего результата.

## Борьба с дисбалансом

### Сбалансируем данные техникой upsampling.

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled

features_upsampled, target_upsampled = upsample(features_train, target_train, 4)
```

####  Проверим модель случайного леса на данных, сбалансированных техникой upsample.

```python
best_result = 0
for depth in range(16, 21):
    for est in range(16, 21):
        for feature in range(1, 6):
            for leaf in range(1, 6):
                rfc_us = RandomForestClassifier(
                    max_depth=depth,
                    n_estimators=est,
                    max_features=feature,
                    min_samples_leaf=leaf,
                    random_state=12345
                )
                rfc_us.fit(features_upsampled, target_upsampled)
                predictions_valid_rfc_us = rfc_us.predict(features_valid)
                probabilities_one_valid_rfc_us = rfc_us.predict_proba(features_valid)[:, 1]
                result = f1_score(target_valid, predictions_valid_rfc_us)
                auc_roc = roc_auc_score(target_valid, probabilities_one_valid_rfc_us)
                matrix = confusion_matrix(target_valid, predictions_valid_rfc_us)
                if result > best_result:
                    best_result = result
                    best_auc_roc = auc_roc
                    best_matrix = matrix
                    best_depth = depth
                    best_est = est
                    best_max_features=feature
                    best_min_samples_leaf=leaf
print(best_result, best_depth ,best_est, best_max_features, best_min_samples_leaf)
print(best_auc_roc)
print(best_matrix)
```

F1-мера этой модели выросла по сравнению с моделью, обученной на несбалансированных данных, теперь она равна 0.6358511837655016 при гиперпараметрах depth = 17, n_estimators = 20, max_features = 2, min_samples_leaf = 5.

Значение ROC-AUC равно 0.8610798780290305, тоже выросло относительно предыдущего случайного леса.

Матрица ошибок:

    TN:1395 FP:198
    FN:125  TP:282

### Сбалансируем данные техникой downsampling.

У нас в распоряжении достаточно данных, чтобы попробовать применить к ним технику downsampling.

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)
    
    return features_downsampled, target_downsampled

features_downsampled, target_downsampled = downsample(features_train, target_train, 0.25)
```

#### Проверим модель случайного леса на данных, сбалансированных техникой downsample.

```python
best_result = 0
for depth in range(16, 21):
    for est in range(16, 21):
        for feature in range(1, 6):
            for leaf in range(1, 6):
                rfc_ds = RandomForestClassifier(
                    max_depth=depth,
                    n_estimators=est,
                    max_features=feature,
                    min_samples_leaf=leaf,
                    random_state=12345
                )
                rfc_ds.fit(features_downsampled, target_downsampled)
                predictions_valid_rfc_ds = rfc_ds.predict(features_valid)
                probabilities_one_valid_rfc_ds = rfc_ds.predict_proba(features_valid)[:, 1]
                result = f1_score(target_valid, predictions_valid_rfc_ds)
                auc_roc = roc_auc_score(target_valid, probabilities_one_valid_rfc_ds)
                matrix = confusion_matrix(target_valid, predictions_valid_rfc_ds)
                if result > best_result:
                    best_result = result
                    best_auc_roc = auc_roc
                    best_matrix = matrix
                    best_depth = depth
                    best_est = est
                    best_max_features=feature
                    best_min_samples_leaf=leaf
print(best_result, best_depth ,best_est, best_max_features, best_min_samples_leaf)
print(best_auc_roc)
print(best_matrix)
```

F1-мера этой модели выросла по сравнению с моделью, обученной на несбалансированных данных, но она ниже, чем мера модели, обученной на данных сбалансированных техникой upsample, теперь она равна 0.6079545454545454 при гиперпараметрах depth = 16, n_estimators = 19, max_features = 5, min_samples_leaf = 5.
