Если пропуски в колонке MultipleLines из phone, значит клиент не пользуется услугами телефонной связи, заменим здесь пропуски на 'no_phone'

```python
data['MultipleLines'] = data['MultipleLines'].fillna('no_phone')
data = data.fillna('no_internet')
```

Дубликаты удалим.

```python
data = data.drop_duplicates()
```

## Добавим новые признаки

### Добавим целевой признак

На основе колонки EndDate создадим целевой признак, где клиент не расторг договор на момент выгрузки данных = 0, расторг = 1

```python
date = pd.to_datetime('2020-02-01')
data['Churn'] = data['EndDate'].apply(lambda x: 0 if x == date else 1)
data.head()
```

Посмотрим на баланс классов

```python
class_frequency = data['Churn'].value_counts(normalize=True)
print(class_frequency)
class_frequency.plot(kind='bar')
```

### Посчитаем длительность договора в днях

```python
data['DurationDays'] = data['EndDate'] - data['BeginDate']
data['DurationDays'] = data['DurationDays'].dt.days.astype(int)
data.head()
```

```python
data['DurationMonths'] = (data['EndDate'] - data['BeginDate'])/np.timedelta64(1, 'M')
data['DurationMonths'] = data['DurationMonths'].astype(int)
data.head()
```

Удалим колонки BeginDate и EndDate, чтобы избежать утечки информации

```python
data = data.drop(['BeginDate', 'EndDate'], axis = 1)
data.head()
```

### Посмотрим, как менялись для клиентов ежемесячные расходы

```python
def payment_changes (row):
    try:    
        if row['TotalCharges'] / row['DurationMonths'] == row['MonthlyCharges']:
            return 'no_changes'
        if row['TotalCharges'] / row['DurationMonths'] < row['MonthlyCharges']:
            return 'increased'
        if row['TotalCharges'] / row['DurationMonths'] > row['MonthlyCharges']:
            return 'reduced'
    except:
        return 'no_changes'
```

```python
data['PaymentChanges'] = data.apply(payment_changes, axis=1)
data.head()
```

```python
data = data.drop(['DurationMonths'], axis=1)
```

### Посмотрим, пользуется ли клиент только услугами телефонной связи, только интернет услугами или обоими сразу

```python
def service_category(row):
    if row['InternetService'] != 'no_internet' and row['MultipleLines'] != 'no_phone':
        return 'both'
    if row['InternetService'] == 'no_internet' and row['MultipleLines'] != 'no_phone':
        return 'phone'
    if row['InternetService'] != 'no_internet' and row['MultipleLines'] == 'no_phone':
        return 'internet'
    if row['InternetService'] == 'no_internet' and row['MultipleLines'] == 'no_phone':
        return 'none'
```

```python
data['ServiceCat'] = data.apply(service_category, axis=1)
data.head()
```

### Посчитаем количество интернет-услуг для каждого интернет-клиента

```python
def service_count(row):
    count = 0
    if row['InternetService'] != 'no_internet':
        if row['OnlineSecurity'] == 'Yes':
            count += 1
        if row['OnlineBackup'] == 'Yes':
            count += 1
        if row['DeviceProtection'] == 'Yes':
            count += 1
        if row['TechSupport'] == 'Yes':
            count += 1
        if row['StreamingTV'] == 'Yes':
            count += 1
        if row['StreamingMovies'] == "Yes":
            count += 1
        return count
    else:
        return 'no_internet'
```

```python
data['ServiceCount'] = data.apply(service_count, axis=1)
data.head()
```

## Подготовка данных для обучения моделей

Фиксируем параметр random_state

```python
RS = 131123
```

### Выделим целевой признак

```python
features = data.drop(['Churn'], axis=1)
target = data['Churn']
```

### Разделим данные на обучающую и тестовую выборки

```python
features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.25, random_state=RS, stratify=target
)

print('Размеры выборок:')
print(f'train = {features_train.shape[0]/len(data):.2%}')
print(f'test = {features_test.shape[0]/len(data):.2%}')
print()
print('Доля ушедших клиентов в выборках:')
print(f'train = {target_train[target_train == 1].sum()/len(data)/0.75:.2%}')
print(f'test = {target_test[target_test == 1].sum()/len(data)/0.25:.2%}')
```

Выборки разделены в соотношении 3:1 и баланс классов в выборках такой же, как в изначальном целевом признаке

### Закодируем категориальные признаки

```python
cat_columns = ['Type', 'PaperlessBilling', 'PaymentMethod', 'SeniorCitizen',
               'Partner', 'Dependents', 'InternetService', 'OnlineSecurity',
               'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
               'StreamingMovies', 'MultipleLines', 'PaymentChanges', 'ServiceCat',
               'ServiceCount']

for col in cat_columns:
    features_train[col] = features_train[col].astype('str')
    
for col in cat_columns:
    features_test[col] = features_test[col].astype('str')
```

```python
transformer = make_column_transformer((OneHotEncoder(drop='first'), cat_columns), remainder='passthrough')

features_train_ohe = pd.DataFrame(transformer.fit_transform(features_train),
                              columns=transformer.get_feature_names_out(),
                              index=features_train.index
                             )
features_test_ohe = pd.DataFrame(transformer.transform(features_test),
                             columns=transformer.get_feature_names_out(),
                             index=features_test.index
                            )
features_train_ohe.head()
```

### Масштабируем количественные признаки

```python
num_columns = ['MonthlyCharges', 'TotalCharges', 'DurationDays']

pd.options.mode.chained_assignment = None

scaler = StandardScaler()
features_train_ohe[num_columns] = scaler.fit_transform(features_train[num_columns])
features_test_ohe[num_columns] = scaler.transform(features_test[num_columns])
features_train_ohe.head()
```

## Обучение моделей

### Обучим модель логистической регрессии

```python
model_lr = LogisticRegression(class_weight='balanced', random_state=RS)
params = {'fit_intercept': [True, False]}

grid_lr = RandomizedSearchCV(model_lr,
                             params,
                             n_iter=2,
                             cv=5,
                             n_jobs=-1,
                             scoring='roc_auc',
                             random_state=RS)
grid_lr.fit(features_train_ohe, target_train)
roc_auc_lr = grid_lr.best_score_

print('Лучшие параметры: ', grid_lr.best_params_)
print('AUC-ROC', roc_auc_lr)
```

### Обучим модель случайного леса

```python
model_rf = RandomForestClassifier(class_weight='balanced', random_state=RS)
params = {'max_depth': [300, 350, 400, 450, 500],
          'n_estimators': [10, 20, 30, 40, 50]
}

grid_rf = RandomizedSearchCV(model_rf,
                             params,
                             n_iter=10,
                             cv=5,
                             n_jobs=-1,
                             scoring='roc_auc',
                             random_state=RS)
grid_rf.fit(features_train_ohe, target_train)
roc_auc_rf = grid_rf.best_score_

print('Лучшие параметры: ', grid_rf.best_params_)
print('AUC-ROC', roc_auc_rf)
```

### Обучим модель LightGBM

```python
model_lgb = lgb.LGBMClassifier(class_weight='balanced', random_state=RS)
params = {
    'max_depth': [2, 4, 6, 8, 10],
    'num_leaves': [10, 20, 30, 40, 50],
    'num_iterations': [100, 200, 500, 1000],
    'n_estimators': [100, 200, 300, 400, 500],
    'learning_rate': [0.03, 0.05, 0.1, 0.2, 0.5]
}

grid_lgb = RandomizedSearchCV(model_lgb,
                              params,
                              n_iter=20,
                              cv=5,
                              scoring='roc_auc',
                              n_jobs=-1,
                              random_state=RS
                             )
grid_lgb.fit(features_train_ohe, target_train)
roc_auc_lgb = grid_lgb.best_score_

print('Лучшие параметры: ', grid_lgb.best_params_)
print('AUC-ROC', roc_auc_lgb)
```

## Проверим лучшую модель на тестовой выборке

Лучше всего себя показала модель LightGBM, проверим ее

```python
best_model = lgb.LGBMClassifier(
    class_weight='balanced',
    num_iterations=500,
    max_depth=2,
    learning_rate=0.5,
    num_leaves=20,
    n_estimators=400,
    random_state=RS
)
best_model.fit(features_train_ohe, target_train)
best_pred_proba = best_model.predict_proba(features_test_ohe)[:, 1]
best_pred = best_model.predict(features_test_ohe)
best_roc_auc = roc_auc_score(target_test, best_pred_proba)
best_accuracy = accuracy_score(target_test, best_pred)
print('AUC-ROC на тестовой выборке: ', best_roc_auc)
print(confusion_matrix(target_test, best_pred))
print('Accuracy на тестовой выборке', best_accuracy)
```
